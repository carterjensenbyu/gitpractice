```python
#must install meteostat using terminal command- pip install meteostat
#importing packages
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
from meteostat import Daily, Point, units
from meteostat.series.convert import convert

DFW = Point(32.89595000000003,-97.03719999999998)
start = datetime(2022,1,1)
end = datetime(2022,1,31)

data = Daily(DFW,start,end)
data = data.convert(units.imperial)
data = data.fetch()

fig = data.plot(y = ['tavg', 'tmin', 'tmax'], kind = 'line',legend=True,xlabel='Dates',ylabel='Daily Temperatures (°F)',title='Daily Temps at DFW, Jan 2022',figsize=(12,6))
plt.savefig('dfwtemps.png')
plt.show()
```
